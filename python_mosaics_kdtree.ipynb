```python
import os
from PIL import Image
from scipy import spatial
import numpy as np
import random
from PIL import Image 
Image.MAX_IMAGE_PIXELS = 100000000000 

main_photo_path = "kobe-bryant-2.jpg"
tile_photos_path = '/Users/norahc/PycharmProjects/Computational_photography/Final_project/datasets_new/'
tile_size = (20, 20)


def getFileList(dir, Filelist, ext=None):
    newDir = dir
    if os.path.isfile(dir):
        if ext is None:
            Filelist.append(dir)
        else:
            if ext in dir[-3:]:
                Filelist.append(dir)
        
    elif os.path.isdir(dir):
        for s in os.listdir(dir):
            newDir = os.path.join(dir, s)
            getFileList(newDir, Filelist, ext)
            
    return Filelist

tile_paths = getFileList(tile_photos_path, [], 'png')
print('I have found '+str(len(tile_paths))+' images\n')



tiles = []
for path in tile_paths:
    try:
        tile = Image.open(path)
        tile = tile.convert("RGB")
        tile = tile.resize(tile_size)
        tiles.append(tile)
    except IOError:
        continue


colors = []
for tile in tiles:
    mean_color = np.array(tile).mean(axis=0).mean(axis=0)[0:3]
    colors.append(mean_color)



main_photo = Image.open(main_photo_path)
main_photo = main_photo.convert("RGB")

width = int(np.round(main_photo.size[0] / tile_size[0]))
height = int(np.round(main_photo.size[1] / tile_size[1]))

resized_photo = main_photo.resize((width, height))

tree = spatial.KDTree(colors, leafsize=6000)



closest_tiles = np.zeros((width, height), dtype=np.uint32)
used = []
index = 0
for i in range(width):
    for j in range(height):
        pixel = resized_photo.getpixel((i, j))
        closest = tree.query(pixel, k=60)
        c = np.array(closest)
        p = 1
        while closest[1][p] in used:
            p = p + 1
            if p == 60:
                used = []
                p = 1
        used.append(closest[1][p])
        
        closest_tiles[i, j] = closest[1][p]
        




output = Image.new('RGB', main_photo.size)


for i in range(width):
    for j in range(height):
        x, y = i*tile_size[0], j*tile_size[1]
        index = closest_tiles[i, j]
        output.paste(tiles[index], (x, y))

output.save("t3_mosaics.png")
```

```python
def ResizeImage(filein, fileout, scale=1):
    """
    change the equality of the image
    :param filein: input image
    :param fileout: output image
    :param scale: scale the image
    :return:
    """
    try:
        img = Image.open(filein)
        width = int(img.size[0] * scale)
        height = int(img.size[1] * scale)
        type = img.format
        out = img.resize((width, height), Image.ANTIALIAS)
        # second parameter：
        # Image.NEAREST ：low equality
        # Image.BILINEAR：Bilinear
        # Image.BICUBIC ：Cubic spline interpolation
        # Image.ANTIALIAS：high equality
        out.save(fileout, type)
    except IOError:
        pass
ResizeImage('t7_mosaics_grid_30.png', 't7_high_resolution_grid_30.png', scale=12)
```

```python
for l in os.listdir('datasets_new/'):
    path = os.path.join('datasets_new/', l)
    if path == 'datasets_new/.DS_Store':
        continue
    ResizeImage(path, path, scale=3)
```
